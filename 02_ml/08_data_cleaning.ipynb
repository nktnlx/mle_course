### Task 4

Fill missing values in `distance_km` column.

```python
groupped_means = data.groupby(['vendor_id', 'store_and_fwd_flag', 'traffic_jam',
              'month_2', 'month_3', 'month_4', 'month_5', 'month_6'])['distance_km'].transform('mean')
```

```python
data['distance_km'] = data['distance_km'].fillna(groupped_means)
```

### Task 5

Build a Linear Regression model using cross validation technique and print out MSLE you've got. 

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_validate


X = data.drop('log_trip_duration', axis=1)
y = data['log_trip_duration']

selector = KFold(n_splits=4, shuffle=True, random_state=33)
model = LinearRegression()
metric = 'neg_mean_squared_error'

result = cross_validate(model, X, y, scoring=metric, cv=selector, return_train_score=True)
cross_val_error = np.mean(result['test_score']) * -1

print(f'MSLE: {round(cross_val_error, 3)}')
```

### Task 8. 

As we can see from the graphs above that we have some outliers in our data. Let's get rid of them.

```python
lower_boundary = np.quantile(y, q=0.005)
upper_boundary = np.quantile(y, q=0.995)
```

```python
data_c = data.copy(deep=True)
```

```python
y_c = data_c[['log_trip_duration']]
```

```python
lower_boundary
```

```python
upper_boundary
```

```python
def rem_outliers(x):
    if x > 3.9318256327243257 and x < 8.32845106681936:
        return 0
    else:
        return 1
```

```python
y_c['is_outlier'] = y_c.log_trip_duration.apply(rem_outliers)
```

```python
data_c = data_c.join(y_c[['is_outlier']])
```

```python
X_new = data_c[data_c['is_outlier'] == 0].drop(['is_outlier', 'log_trip_duration'], axis=1)
```

```python
X_new
```

```python
y_new = data_c[data_c['is_outlier'] == 0]['log_trip_duration']
```

```python
y_new.shape
```

### Task 9. 

Build a Linear Regression model using cross validation technique on data having no outliers and print out MSLE you've got. 

```python
model = LinearRegression()
metric = 'neg_mean_squared_error'

result = cross_validate(model, X_new, y_new, scoring=metric, cv=selector, return_train_score=True)

cross_val_error = np.mean(result['test_score']) * -1

print(f'MSLE: {round(cross_val_error, 3)}')
```
